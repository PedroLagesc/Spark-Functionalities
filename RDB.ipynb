```
# Import the modules
from pyspark.sql import SparkSession
from pyspark.sql.types import *
```

```
# Let´s create a RDB
num = sc.parallelize([1,2,3,4,5,6,7,8,9])
```

```
# Let´s select the first 6 numbers
num.take(6)
```

```
# Showing the top 3 values (>)
num.top(3)
```

```
# Filtering the values higher than 4
filt = num.filter (lambda filt : filt > 4)
filt.collect()
```

```
samp = num.sample(True , 0.5 , 1)
samp.collect()
```

```
# Using Map to create a lambda function to multiply the values by 2
mapa = num.map (lambda mapa : mapa * 2)
mapa.collect()
```

```
num_2 = sc.parallelize([10,11,12,13,14,15])
```

```
# Let´s bound the 2 RDBs
bound = num.union(num_2)
bound.collect()
```

```
cartesiano = num.cartesian(num_2)
cartesiano.collect()
```

```
# Counting how many times the values appear in the RDB
cartesiano.countByValue()
```

```
# Build a RDB with tuples
buy = sc.parallelize([(1,200), (2,300) , (3, 1000), (4, 800) , (5,900)])
buy.collect()
```

```
# Costumers Ids
keys = buy.keys()
keys.collect()
```

```
# Values of the purchases
values = buy.values()
values.collect()
```

```
buy.countByKey()
```

```
mysum = buy.mapValues(lambda mysum : 1 + mysum)
mysum.collect()
```
