```
# Loading the file from the Defined Path where Header = True
df_ibov = spark.read.format("csv").option("header", True ).load("/learn/b3_stocks")
```

```
# Print the data frame - > String fields
df_ibov.show()
df_ibov.printSchema()
```

```
# LetÂ´s CAST our Data Types
df_ibov = df_ibov.select (
  col("datetime").cast("date"),
  col("ticker").cast("string"),
  col("open").cast("double"),
  col("close").cast("double"),
  col("high").cast("double"),
  col("low").cast("double"),
  col("volume").cast("double")
  
  )
df_ibov.printSchema()
```

```
# LetÂ´s save the new data frame
df_ibov.write.format("csv").mode("overwrite").option("header", "true").save("/learn/b3_stocks_converted")
```

```
# Filtering to display the data required by the where function
df_ibov.select("ticker" , "high" , "low").where(col("high") > 100).show()
```

```
# Using more conditions in the Where Function
df_ibov.select("ticker" , "high" , "low").where((col("high") > 200) & (col("low") > 100 )).show()
```

```
# Changing the column name and showing it
df_new = df_ibov.withColumnRenamed("datetime" , "date")
df_new.columns
```

```
# Changing the Date Type - > To Timestamp
df_new_2 = df_ibov.withColumn("datetime", to_timestamp(col("datetime"), "yyyy-MM-dd"))
df_new_2.select("datetime").printSchema()
df_new_2.select("datetime").show()
```

```
df_new_2.select("datetime", "ticker" , "open").groupBy("datetime").count().show()
```

```
df_new_2.select("datetime", "ticker" , "open").orderBy("datetime").show()
```
