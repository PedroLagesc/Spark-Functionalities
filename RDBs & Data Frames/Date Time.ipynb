```
# Importing the lybraries
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
# LetÂ´s create dataframes as strings
df_data_1 = spark.createDataFrame(['2018-01-01 00:00:00+00:00', '2018-07-05 01:00:00+00:00','2018-09-06 02:00:00+00:00'], "string").toDF("dates")
df_data_2 = spark.createDataFrame(['2017-12-31 16:00:00', '2017-12-15 17:00:00','2017-12-05 18:00:00'], "string").toDF("dates")
df_data_3 = spark.createDataFrame(['2018-02-01', '2018-11-18', '2018-12-03'], "string").toDF("dates")
```

```
# Displaying our DFs
display(df_data_1)
display(df_data_2)
display(df_data_3)
```

```
df_data_1.createOrReplaceTempView("date_1")
df_data_2.createOrReplaceTempView("date_2")
df_data_3.createOrReplaceTempView("date_3")
```

```
%sql
SELECT
  TO_TIMESTAMP(dates) dates
FROM date_1
```

```
df_sql_data_1 = df_data_1

df_sql_data_1 = spark.sql("""
  SELECT
    TO_TIMESTAMP(dates) dates
  FROM date_1
""")
df_sql_data_1.display()

df_sql_data_2 = df_data_2

df_sql_data_2 = spark.sql("""
  SELECT
    CAST(dates AS DATE) dates
  FROM date_2
""")
df_sql_data_2.display()
```

```
%sql
SELECT
  --TO_DATE(dates,"yyyy-MM-dd") date_brazil
  TO_TIMESTAMP(dates,"yyyy-MM-dd") timestamp_brazil
FROM date_3
```

```
df_spark_data_1 = df_data_1
df_spark_data_1 = df_spark_data_1.withColumn( "dates", to_timestamp("dates") ) 
df_spark_data_1.printSchema()
```

```
df_spark_data_2 = df_data_2
df_spark_data_2 = df_spark_data_2.withColumn("dates" , to_date("dates"))
df_spark_data_2.printSchema()
```

```
df_spark_data_3 = df_data_3
df_spark_data_3 = df_spark_data_3.withColumn("dates" , to_date("dates" , "yyyy-MM-dd"))
df_spark_data_3.display()
df_spark_data_3.printSchema()
```
