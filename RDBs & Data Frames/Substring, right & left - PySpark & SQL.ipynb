```
# Importing the lybraries
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
# Read the data
df_car = spark.read.format("csv").option ("header" , True).load("/learn/modelo_carro")
df_car.display()
```

```
df_car.createOrReplaceTempView("cars")
```

```
%sql
-- SUBSTRING (column , initial position , number of chars to run) --
SELECT
  modelo_carro,
  SUBSTRING(modelo_carro, 2 , 3),
  LEFT(modelo_carro, 2),
  RIGHT(modelo_carro, 3)
FROM cars
```

```
df_car.createOrReplaceTempView("cars")
df_car_sql = spark.sql ("""
  SELECT
    modelo_carro,
    SUBSTRING(modelo_carro, 2 , 3) mod_car,
    LEFT(modelo_carro, 2) first_two,
    RIGHT(modelo_carro, 3) last_three
  FROM cars
""")

df_car_sql.display()
```

```
df_car_spark = df_car
df_car_spark = df_car_spark.withColumn (
  "modelo_substr", substring("modelo_carro", 2 , 4)
).withColumn(
  "modelo_left", expr("LEFT(modelo_carro, 2)") 
).withColumn(
  "modelo_right", expr("RIGHT(modelo_carro, 4)")
)

display(df_car_spark)
```
