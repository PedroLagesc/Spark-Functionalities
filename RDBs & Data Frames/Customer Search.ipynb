```
# Import the modules
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
# Loading the data
customers = spark.read.format("parquet").load("/learn/Clientes")
sales  = spark.read.format("parquet").load("/learn/Vendas")
```

```
# We will verify the schema type
sales.printSchema()
sales.show()
```

```
customers.printSchema()
customers.show()
```

```
# Searching for the status and the state of our customers
customers.select("Cliente", "Estado", "Status").show()
```

```
search_1 = customers.select("Cliente", "Estado", "Status").show()
```

```
# Searching for only the Gold & Platinum Customers
customers.select("*").where( (col("Status")=="Platinum") | (col("Status")=="Gold") ).show()
```

```
# Store the search in a new Data Frame
search_2 = customers.select("*").where( (col("Status")=="Platinum") | (col("Status")=="Gold") ).show()
```

```
# We will join the both Datasets to perform our search
sales.join(customers, sales.ClienteID == customers.ClienteID).groupBy(customers.Status).agg(sum("Total")).orderBy(col("sum(Total)").desc()).show()
```

```
# Grouping our customers by their status , adding the total purchases , ordering by the Total purchases in a decreasig order
search_3 = sales.join(customers, sales.ClienteID == customers.ClienteID).groupBy(customers.Status).agg(sum("Total")).orderBy(col("sum(Total)").desc())
search_3.show()
```
