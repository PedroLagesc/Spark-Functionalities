```
# Import the modules
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
# Loading the file from the Defined Path where Header = True
df_ibov = spark.read.format("csv").option("header", True ).load("/learn/b3_stocks_converted")
```

```
# Print the data frame - > String fields
df_ibov.printSchema()
df_ibov.show()
```

```
# LetÂ´s CAST our Data Types
df_ibov = df_ibov.select (
  col("datetime").cast("date"),
  col("ticker").cast("string"),
  col("open").cast("double"),
  col("close").cast("double"),
  col("high").cast("double"),
  col("low").cast("double"),
  col("volume").cast("double")
  
  )
df_ibov.printSchema()
```

```
df_ibov.show()
```

```
df_ibov.orderBy("open").show()
```

```
# Ordering by decreasing order
df_ibov.orderBy(col("open").desc()).show()
```

```
# Ordering the Ticker by the decreasing order of the higher stock value
df_ibov.orderBy(col("ticker").desc(),col("high").desc() ).show()
```

```
# Grouping the tickers by the total volume negotiated - > using groupBy & agg function
df_ibov.groupBy("ticker").agg(sum("volume")).show()
```

```
df_ibov.groupBy("ticker").agg(sum("volume")).orderBy(col("sum(volume)").desc()).show()
```

```
# Looking for the Petrorio negotiations 
df_ibov.filter(col("ticker")=="PRIO3").show()
```
