```
churn = spark.read.format("csv").option("header", True ).load("/learn/Churn")
churn.show()
```

```
# Write the data as a partition
churn.write.partitionBy("Geography").saveAsTable("churn_location")
```

```
# checking the table
spark.sql(" SELECT * FROM churn_location ").show(10)
```

```
# Using the bucket by function
churn.write.bucketBy(2 , "Geography").saveAsTable("churn_location_bucket")
# checking the table
spark.sql(" SELECT * FROM churn_location_bucket ").show(10)
```
