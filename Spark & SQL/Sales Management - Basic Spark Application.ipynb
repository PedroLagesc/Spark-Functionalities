```
# Import the modules
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
# Loading the data
customers = spark.read.format("parquet").load("/learn/Clientes")
sales  = spark.read.format("parquet").load("/learn/Vendas")
products = spark.read.format("parquet").load("/learn/Products")
itens = spark.read.format("parquet").load("/learn/ItemsSold")
vendors = spark.read.format("parquet").load("/learn/Vendors")
```

```
# In the next steps we will build a Management Table relating the paramount data
sales.join(customers, sales.ClienteID == customers.ClienteID, "inner" ).show()
```

```
management_table = sales.join(customers, sales.ClienteID == customers.ClienteID, "inner" )
management_table.show()
```

```
management_table = management_table.join(itens, management_table.VendasID == itens.VendasID, "left" )
management_table.show()
```

```
management_table = management_table.join(vendors , management_table.VendedorID == vendors.VendedorID, "left")
management_table.show()
```

```
# So in the end we will have the following Management Table
management_table.printSchema()
management_table.show()
```
