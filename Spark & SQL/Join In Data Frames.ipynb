```
# Import the modules
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
# Loading the data
customers = spark.read.format("parquet").load("/learn/Clientes")
sales  = spark.read.format("parquet").load("/learn/Vendas")
```

```
# We will verify the schema type
sales.printSchema()
sales.show()
```

```
customers.printSchema()
customers.show()
```

```
sales.write.saveAsTable("sAles")
```

```
# Using Spark SQL to do a Join
spark.sql("SELECT sAles .*, customers_new.ClienteID FROM customers_new INNER JOIN sAles ON (sAles.ClienteID = customers_new.ClienteID)").show()
```

```
# Doing a Join With Spark API
sales.join(customers, sales.ClienteID == customers.ClienteID, "inner" ).select("VendedorID", "Total").show()
```

```
sales.join(customers, sales.ClienteID == customers.ClienteID, "right" ).select("Data","VendedorID", "Total").show()
# sales.join(customers, sales.ClienteID == customers.ClienteID, "left" ).select("Data","VendedorID", "Total").show()
```
