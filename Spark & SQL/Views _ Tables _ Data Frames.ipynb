```
# Import the modules
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
# Loading the data
sales  = spark.read.format("parquet").load("/learn/Vendas")
```

```
# Creating a Temporary View
sales.createOrReplaceTempView("sales_view_1")
```

```
# Select Everything from the Temp View
spark.sql ("SELECT * FROM sales_view_1").show()
```

```
# Creating a Global Temporary View
sales.createOrReplaceGlobalTempView("sales_view_2")
```

```
# Select Everything from the Global Temp View
spark.sql ("SELECT * FROM global_temp.sales_view_2").show()
```

```
# Creating a temp view using primary SQL Commands
spark.sql("CREATE OR REPLACE TEMP VIEW SALES_VIEW AS SELECT * FROM sales_view_1")
```

```
spark.sql("SELECT * FROM SALES_VIEW").show()
```

```
# Creating a global temp view using primary SQL Commands
spark.sql("CREATE OR REPLACE GLOBAL TEMP VIEW SALES_VIEW_2 AS SELECT * FROM sales_view_1")
```

```
spark.sql("SELECT * FROM global_temp.SALES_VIEW_2").show()
```

```
# Selecting multiple columns and applying the WHERE
spark.sql("SELECT ClienteID, Total FROM sales_view_1 WHERE Total > 10000").show()
```

```
# Combine the Sum , Group By and Order By functions 
spark.sql("SELECT ClienteID, SUM(Total) FROM SALES_VIEW group by ClienteID ORDER BY 2 desc").show()
```
