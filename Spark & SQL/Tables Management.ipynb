```
# Import the modules
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
# Loading the data
customers = spark.read.format("parquet").load("/learn/Clientes")
sales  = spark.read.format("parquet").load("/learn/Vendas")
```

```
# Creating an External Table
customers.write.option("path", "/learn/Customers" ).saveAsTable("customers_new")
```

```
spark.sql("SELECT * FROM customers_new ").show()
```

```
# Verify if the Table is a External Table
spark.sql("SHOW CREATE TABLE customers_new ").show(truncate=False)
```

```
# Showing the Tables in the database
spark.catalog.listTables()
```
