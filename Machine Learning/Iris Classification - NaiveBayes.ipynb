```
# Importing the main modules
from pyspark.ml.feature import RFormula
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.classification import NaiveBayes
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```
iris = spark.read.format("csv").option("header", True ).option("inferSchema", True).load("/learn/Iris")
```

```
iris.printSchema()
iris.show()
```

```
# LetÂ´s apply the R Formula to define the dependents variables
formula = RFormula(formula = "class ~." , featuresCol="features" , labelCol="label", handleInvalid="skip")
```

```
# Fitting and transforming the iris
iris_transformed = formula.fit(iris).transform(iris).select("features","label")
```

```
# Check the new columns of features and label
iris_transformed.show(10)
```

```
# Random Split to define the % of variables to be Trained and Tested
iris_train , iris_test = iris_transformed.randomSplit([.7 , .3 ])
```

```
# Define the Naive Bayes Object
mynaive = NaiveBayes(featuresCol="features",labelCol="label")
```

```
# Naive Bayes Model
model = mynaive.fit(iris_train)
# Predict the Train Model
predicted = model.transform(iris_test)
```

```
# predicted.show(truncate=False)
predicted.show()
```

```
# Creating the Evaluator Object using the Multiclass Classification Evaluator
evaluator = MulticlassClassificationEvaluator(predictionCol="prediction" , labelCol="label" , metricName="accuracy")
```

```
# LetÂ´s see the reult of the Evaluation of our predicted model
result = evaluator.evaluate(predicted)
print(result)
# As we can see it is a great result = > Model can be destined to production
```
